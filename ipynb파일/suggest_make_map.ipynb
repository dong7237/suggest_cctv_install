# 선정한 후보 격자,열에 대한 지도 시각화
- 인사이트를 활용해 지도 내 세부 설치 위치 선정

```python
import pandas as pd
import numpy as np
```

```python
cctv_df = pd.read_csv('격자화_cctv장소.csv')
infra_df = pd.read_csv('격자화_인프라.csv')
light_df = pd.read_csv('격자화_가로등.csv')
illigal_df = pd.read_csv('격자화_불법주정차단속.csv')

cand_col = ['가로등','대규모점포','체육도장','골프연습장','미용','휴게음식점']
```

```python
import pickle
with open('grid_candidate.pkl', 'rb') as file:
    grid_candidate = pickle.load(file)
# 후보격자 불러오기
```

```python
with open('columns_candidate.pkl', 'rb') as file:
    cand_col = pickle.load(file)
# 후보 열 불러오기
```

```python
infra_df =infra_df[infra_df['index_right'].isin(np.float64(grid_candidate['grid_cand']))]
light_df =light_df[light_df['index_right'].isin(np.float64(grid_candidate['grid_cand']))]
illigal_df = illigal_df[illigal_df['index_right'].isin(np.float64(grid_candidate['grid_cand']))]
# 모델이 선정한 격자만을 추출
infra_df = infra_df[infra_df['영업상태명']!='폐업']
infra_df = infra_df[infra_df['정보'].isin(cand_col)]
infra_df.rename(columns={'좌표정보(X)':'경도'},inplace=True)
infra_df.rename(columns={'좌표정보(Y)':'위도'},inplace=True)


infra_df = infra_df.reset_index()
cctv_df = cctv_df.reset_index()
light_df = light_df.reset_index()
illigal_df = illigal_df.reset_index()
```

```python
import folium
```

```python
def add_circles(map_obj, dataframe):
    for index, row in dataframe.iterrows():
        folium.Circle(
            location=[row['위도'], row['경도']],
            radius=100,
            color='blue',
            fill=True,
            fill_color='blue',
            fill_opacity=0.1
        ).add_to(map_obj)
```

```python
def add_data_points(map_obj, dataframe, color):
    for index, row in dataframe.iterrows():
        if color == 'blue':
            radius = 4
            opa = 0.7
        if color == 'green':
            radius = 5
            opa = 0.7
        if color == 'red':
            radius = 10
            opa = 0.01
        if '위도' in row and '경도' in row:
            folium.CircleMarker(
                location=[row['위도'], row['경도']],
                radius=radius,
                color=color,
                fill=True,
                fill_color=color,
                fill_opacity=opa
            ).add_to(map_obj)
            print(color)

```

## 각 인프라 별 1대 설치가 목표이므로 각 격자마다 시각화가 필요함.
- 개별 격자 시각화를 통해 격자 별 설치 장소 1곳 결정

```python
grid_candidate
```

```python
for i in grid_candidate['grid_cand']:
    z = float(i)
    m =  folium.Map(location=[37.541205,127.069405], zoom_start=14)    
    add_data_points(m, infra_df[infra_df['index_right']==z], 'green')  
    add_data_points(m, illigal_df[illigal_df['index_right']==z], 'red')    
    add_circles(m, cctv_df[cctv_df['index_right']==z])
    add_data_points(m, cctv_df[cctv_df['index_right']==z], 'blue')    

    updated_map_path = f"folium_map_with_data_points({z}).html"
    m.save(updated_map_path)
```
